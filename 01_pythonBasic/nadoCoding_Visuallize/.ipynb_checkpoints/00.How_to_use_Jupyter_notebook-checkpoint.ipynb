# Jupyter Notebook


주피터 노트북은 웹 브라우저 상에서 개발을 할 수 있는 도구이며, 코드를 Cell 단위로 묶어서 실행하고 그래프나 표, 그리고 이미지나 영상 등을 쉽게 볼 수 있어서 특히 데이터 관련 작업을 할 때 많이 활용합니다.

> 교육을 위한 강의 노트로도 아주 훌륭합니다!

[자주 사용하는 단축키]
- window 기준, mac에서는 상이할 수 있음
- Help -> Keyboard shortcuts에서 단축키 확인 가능
    - Command Mode / Edit Mode로 나뉘어 있음

Ctrl + Enter : 현재 Cell 수행  
Shift + Enter : 현재 Cell 수행 후 다음 Cell 선택  
Alt + Enter : 현재 Cell 수행 후 다음 위치에 Cell 삽입  
Esc : 편집모드 -> 명령모드  
b : (명령모드에서 사용)현재 셀 밑에 셀 추가  
d, d : (명령모드에서 사용)현재 셀 삭제  
Shift + L : (명령모드에서 사용)현재 셀 라인 넘버 표시
m : (명령모드에서 사용)현재 셀 양식 -> Markdown
y : (명령모드에서 사용)현재 셀 양식 -> Code
tab : 들여쓰기  
- 모듈에서 어떤 함수를 제공하는지 모를 때, tab 사용하면 목록을 표시  

Ctrl + / : 주석처리  
i, i : 키보드 인터럽트 발생

## 특징

1. 코드를 Cell 단위로 작성 및 실행
1. 이렇게 (1. 1.)로 작성하게되면 자동으로 번호를 순서대로 표시해준다.
1. (1. 2. 3.) 이렇게하면 중간에 삭제한 번호때문에 나머지를 다 바꿔줘야 한다
1. 마크다운을 통한 문서화
1. 그래프나 표 등을 실시간으로 확인
1. html, pdf 등 파일로 저장 가능

## 시각화 예제 2 : 테이블

- pandas는 복잡한 데이터를 분석할 때 아주 유용합니다.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl

df = pd.DataFrame([[38.0, 2.0, 18.0, 22.0, 21, np.nan],[19, 439, 6, 452, 226,232]],
                  index=pd.Index(['Tumour (Positive)', 'Non-Tumour (Negative)'], name='Actual Label:'),
                  columns=pd.MultiIndex.from_product([['Decision Tree', 'Regression', 'Random'],['Tumour', 'Non-Tumour']], names=['Model:', 'Predicted:']))
df.style
```
