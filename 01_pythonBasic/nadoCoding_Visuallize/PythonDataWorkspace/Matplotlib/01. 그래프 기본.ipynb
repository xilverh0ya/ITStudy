# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
print(plt.plot(x, y))
```

```python
# 이 방식으로 하면 불필요한 데이터 표시 없이 깔끔하게 그래프만 출력됨.
plt.plot(x, y)
plt.show()
```

## Title 설정

```python
plt.plot(x, y)
plt.title('Line Graph')
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
```
