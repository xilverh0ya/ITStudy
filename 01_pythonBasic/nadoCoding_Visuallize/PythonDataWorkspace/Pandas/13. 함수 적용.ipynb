# 13. 함수 적용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원 번호')
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

```python
df['키'] = df['키'] + 'cm'
```

- df['키'] 데이터는 int형이고, + 해주려는 'cm'는 str형이기 떄문에 오류가 발생한다

## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm를 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 각 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

- 첫 번째 글자 : 대문자
- 나머지 글자 : 소문자
- 적용 Column : SW특기

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN 이 아닌지
        return lang.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

- str함수에도 capitalize가 있기 떄문에 실전에선 다음과 같이 쓰는게 좋음.

```python
df['SW특기'].str.capitalize()
```
