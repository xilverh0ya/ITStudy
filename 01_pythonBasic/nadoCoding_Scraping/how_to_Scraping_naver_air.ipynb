````python
# selenium_flight.py

# 대기를 위한 time
import time

# 준비
from selenium import webdriver
from selenium.webdriver.common.by import By

# 대기를 위한 WebDriverWait import
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC # 기다리는 조건을 뭘로 할 지 정해줌

# 값을 기다릴 때마다 WebDriverWait을 호출해야하므로 함수로 정의해버리자.
def wait_until(xpath_str):
    WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, xpath_str)))

# 크롬 창을 열기
browser = webdriver.Chrome()

# 브라우저 최대화
browser.maximize_window()

# 열린 크롬창에서 해당 사이트 접속
url = 'https://flight.naver.com/' # 네이버 항공권 사이트
browser.get(url)

# 가는 날 버튼을 찾기(Chrome의 개발자 모드 이용)
begin_date = browser.find_element(By.XPATH, '//button[text() = "가는 날"]') # // : HTMl문서 전체에서 찾겠다는 뜻

# 가져온 데이터 클릭하기
begin_date.click()

# 열린 '가는 날 선택' 창에서 5월 27일 선택하기
# 가는 날 선택 후 캘린더 창이 뜰 때 까지 기다려줘야 함.
# WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//b[text() = "27"]')))
# time.sleep(1) # 1초 대기
wait_until('//b[text() = "27"]')

## 수 많은 27일 중 첫 번째 27일이 5월 27일이 될 것.
day27 = browser.find_elements(By.XPATH, '//b[text() = "27"]')

# 몇 개의 27일 데이터를 가져왔는 지 확인
len(day27)

# 첫 번째 day27 데이터 클릭
day27[0].click()

# 가는 날 선택이 완료되면 '오는 날 선택' 창이 열림
# WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//b[text() = "30"]')))
wait_until('//b[text() = "30"]')
# '오는 날 선택' 창에서 5월 31일 선택하기
day31 = browser.find_elements(By.XPATH, '//b[text() = "30"]')

# 몇 개의 31일 데이터를 가져왔는 지 확인
# len(day31)

# 첫 번째 31일 데이터 클릭
day31[0].click()


# 도착 부분 클릭하기

# WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//b[text() = "도착"]')))
wait_until('//b[text() = "도착"]')
# 도착 버튼 찾기
arrival = browser.find_element(By.XPATH, '//b[text() = "도착"]')

# 도착 버튼 클릭하기
arrival.click()

# WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//button[text() = "국내"]')))
# time.sleep(1) # 1초 대기
wait_until('//button[text() = "국내"]')

# 국내 버튼 찾고 클릭
domestic = browser.find_element(By.XPATH, '//button[text() = "국내"]')
domestic.click()

# WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//i[contains(text(), "제주국제공항")]')))
# time.sleep(1) # 1초 대기
wait_until('//i[contains(text(), "제주국제공항")]')

# "제주국제공항"이란 텍스트가 포함 된 버튼을 찾아서 클릭
jeju = browser.find_element(By.XPATH, '//i[contains(text(), "제주국제공항")]')
jeju.click()

# 항공권 검색 버튼 찾아서 클릭
search = browser.find_element(By.XPATH, '//span[contains(text(), "항공권 검색")]')
search.click()

# 로딩되는 시간동안 기다려야 함!!
# 값이 표시될 때 까지 기다림 (domestic_Flight_sK0eA result 프레임이 나올 때 까지)
elem = WebDriverWait(browser, 30).until(EC.presence_of_element_located((By.XPATH, '//div[@class = "domestic_Flight__sK0eA result"]')))
print(elem.text)

# 브라우저 종료
browser.quit()
````

- b : b 태그 검색
- text() = "27" : b태그를 가진 텍스트들 중에서 27을 검색
